```python
import warnings
warnings.filterwarnings('ignore')
import sys
sys.path.insert(0,'../')
import numpy as np
import datetime
import matplotlib
%matplotlib inline
import pylab as pl
import pandas as pd
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pydivide
import swia
import utils
from scipy.interpolate import interp1d
```

### Reproducing figure 2 bottom row Halekas et al. 2015

```python
#We use pydivide to download spacecraft ephemeris and magnetic field data
date = '2015-02-27'
tStart = '03:00:00'
tEnd = '05:00:00'
pydivide.download_files(start_date=date,end_date=date,insitu=True,exclude_orbit_file=True)
```

```python
#Read spacecraft ephemeris and magnetic field data
insitu = pydivide.read(date,insitu_only=True)
spAtrbs = ['SZA','ALTITUDE']
magAtrbs = ['MSO_X','MSO_Y','MSO_Z']
velAtrbs = ['HPLUS_FLOW_VELOCITY_MSO_X','HPLUS_FLOW_VELOCITY_MSO_Y','HPLUS_FLOW_VELOCITY_MSO_Z']
dS = insitu['SPACECRAFT'][spAtrbs].join(insitu['MAG'][magAtrbs])
dS.index = pd.to_datetime(dS.index)
dS_Trn = utils.cutDstoRange(dS,date,tStart,tEnd) 
dS_Trn
```

```python
#Download and read SWIA data in cdf format
obsResTyps = ['coarse','fine']
dF = {}
for tyP in obsResTyps:
    dF[tyP] = swia.getData(date,dataCls='%s_svy_3d'%tyP)
```

```python
#Compute penetrating and backscattered proton flux from coarse observations
energyArr = {}
timeArr = {}
forwardFluxes,backwardFluxes,energyArr['coarse'], timeArr['coarse'] = swia.reduceCoarseData(dF['coarse'],date,tStart,tEnd)
```

```python
#Angle integrated spectra
#Solar Wind
swStart = '03:15:00'
swEnd = '03:16:30'
thisPhi = dF['fine']['phi_fine'][...]*swia.degToRad
sW, swTime = swia.getSAIntegratedFlux(dF['fine'],'fine',thisPhi,range(len(thisPhi)),date,swStart,swEnd)
sW = sW.mean(axis=0)
sw_En = np.zeros(48,dtype=np.float)
startId,endId = utils.selCdfTimes(dF['fine'],date,swStart,swEnd)
for i in range(startId,endId):
    sw_En += dF['fine']['energy_fine'][...][dF['fine']['estep_first'][i]:dF['fine']['estep_first'][i]+48]
sw_En /= (endId-startId)
```

```python
#Sheath
stStart = '03:25:00' 
stEnd = '03:30:00'
thisPhi = dF['coarse']['phi_coarse'][...]*swia.degToRad
sT,stTime = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[:7],range(7),date,stStart,stEnd)
sT2,_ = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[7:],range(7,16),date,stStart,stEnd)
sT = sT.mean(axis=0)
sT2 = sT2.mean(axis=0)
sT += sT2
st_En = dF['coarse']['energy_coarse'][:]
```

```python
#forward
spStart = '03:50:00'
spEnd = '04:00:00'
thisPhi = dF['coarse']['phi_coarse'][...]*swia.degToRad
sP,spTime = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[13:16],range(13,16),date,spStart,spEnd)
sP2,_ = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[0:1],range(0,1),date,spStart,spEnd)
sP = sP.mean(axis=0)
sP2 = sP2.mean(axis=0)
sP += sP2
sp_En = dF['coarse']['energy_coarse'][:]
```

```python
#backward
thisPhi = dF['coarse']['phi_coarse'][...]*swia.degToRad
sB,sbTime = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[1:7],range(1,7),date,spStart,spEnd)
sB2,_ = swia.getSAIntegratedFlux(dF['coarse'],'coarse',thisPhi[7:13],range(7,13),date,spStart,spEnd)
sB = sB.mean(axis=0)
sB2 = sB2.mean(axis=0)
sB += sB2
sb_En = dF['coarse']['energy_coarse'][:]
```

```python
#Interpolation of energy bin values to get "nice" yticks
energieS = np.log10(dF['coarse']['energy_coarse'][:])
indiceS = range(len(energieS)) 
intpFn = interp1d(energieS,indiceS)
```

```python
#Main plot
def forceAspect(ax,aspect=1):
    im = ax.get_images()
    extent =  im[0].get_extent() 
    ax.set_aspect(abs((extent[1]-extent[0])/(extent[3]-extent[2]))/aspect)

matplotlib.rc('axes', labelsize=18)
matplotlib.rc('xtick', labelsize=16)
matplotlib.rc('ytick', labelsize=16)
matplotlib.rc('legend', fontsize=10)
gs = gridspec.GridSpec(38,9)
fig = pl.figure(figsize=(14,16))
ax = fig.add_subplot(gs[:6,:4])
ax.plot(dS_Trn.index,dS_Trn['MSO_X'],lw=2,c='b',label='Bx')
ax.plot(dS_Trn.index,dS_Trn['MSO_Y'],lw=2,c='r',label='By')
ax.plot(dS_Trn.index,dS_Trn['MSO_Z'],lw=2,c='y',label='Bz')
ax.set_xticks(dS_Trn.index.values[::150])
ax.set_xticklabels([str(x).split('T')[1][1:5] for x in dS_Trn.index.values[::150]],rotation=30)
ax.set_yticks(np.arange(-100,150,50))
ax.set_yticklabels(np.arange(-100,150,50))
ax.set_ylabel('B (MSO, nT)')
ax.legend(loc=0)

energyLabels_minor = np.arange(1.5,4.3,0.1)
energyTicks_minor = intpFn(energyLabels_minor)
energyLabels_minor = np.power(10.,energyLabels_minor)

energyLabels_major = np.arange(2.0,5.0,1.0)
energyTicks_major = intpFn(energyLabels_major)
energyLabels_major = np.power(10.,energyLabels_major)


ax = fig.add_subplot(gs[8:14,:4])
im = ax.imshow(forwardFluxes,cmap='gnuplot2',vmin=0,vmax=10)
ax.set_yticks(energyTicks_major)
ax.set_yticklabels(energyLabels_major)
ax.set_ylabel('Forward')
ax.set_yticks(energyTicks_minor,minor=True)
xTicks = np.array([x.strftime('%H:%M')[1:] for x in timeArr['coarse']])
ax.set_xticks(range(0,750,90))
ax.set_xticklabels(xTicks[0:750:90],rotation=30)
forceAspect(ax,3.5)
cbar = pl.colorbar(im,cax = fig.add_axes([0.465,0.61,0.015,0.1]),ticks=range(1,10))
cbar.ax.tick_params(labelsize=10)
cbar.ax.set_yticklabels(['$10^%d$'%x for x in range(1,10)])  # vertically oriented colorbar 
cbar.ax.set_xlabel('diff en flux',fontsize=10)
cbar.ax.xaxis.set_label_coords(0.5,-0.15)

ax = fig.add_subplot(gs[16:22,:4])
im = ax.imshow(backwardFluxes,cmap='gnuplot2',vmin=0,vmax=10)
ax.set_yticks(energyTicks_major)
ax.set_yticklabels(energyLabels_major)
ax.set_ylabel('Back')
ax.set_yticks(energyTicks_minor,minor=True)
xTicks = np.array([x.strftime('%H:%M')[1:] for x in timeArr['coarse']])
ax.set_xticks(range(0,750,90))
ax.set_xticklabels(xTicks[0:750:90],rotation=30)
forceAspect(ax,3.5)
cbar = pl.colorbar(im,cax = fig.add_axes([0.465,0.45,0.015,0.1]),ticks=range(1,10))
cbar.ax.tick_params(labelsize=10)
cbar.ax.set_yticklabels(['$10^%d$'%x for x in range(1,10)])  # vertically oriented colorbar 
cbar.ax.set_xlabel('diff en flux',fontsize=10)
cbar.ax.xaxis.set_label_coords(0.5,-0.15)

ax = fig.add_subplot(gs[24:30,:4])
ax.plot(dS_Trn.index,dS_Trn['ALTITUDE'],lw=2,c='b')
ax.set_xticks(dS_Trn.index.values[::150])
ax.set_xticklabels([str(x).split('T')[1][1:5] for x in dS_Trn.index.values[::150]],rotation=30)
ax.set_yscale('log')
ax.set_ylim((80,20000))
ax.set_ylabel('altitude (km)')
ax.set_yticks([100,1000,10000])
ax.set_yticklabels([100,1000,10000])

ax = fig.add_subplot(gs[32:38,:4])
ax.plot(dS_Trn.index,dS_Trn['SZA'],lw=2,c='b')
ax.set_yticks([45,90,135])
ax.set_yticklabels([45,90,135])
ax.set_xticks(dS_Trn.index.values[::150])
ax.set_xticklabels([str(x).split('T')[1][1:5] for x in dS_Trn.index.values[::150]],rotation=30)
ax.set_ylabel('SZA')

ax = fig.add_subplot(gs[5:27,5:])
ax.plot(st_En,sT,lw=2,color='y',label='sheath')
ax.plot(sp_En,sP,lw=2,color='r',label='penetrating')
ax.plot(sb_En,sB,lw=2,color='b',label='backscattered')
ax.plot(sw_En[8:-8],sW[8:-8],lw=2,color='k',label='solar wind')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('Angle Integrated Flux')
pl.legend(loc=0,fontsize=14)
pl.show()
```
